```python
# Imports
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data_dir = "../data/"
brazil_path = os.path.join(data_dir, "Brazil.csv")
usa_path = os.path.join(data_dir, "United States of America.csv")
if not os.path.exists(brazil_path) or not os.path.exists(usa_path):
	print("Please run the data download script first")
	exit()
```

### Read Data

```python
brazil_df = pd.read_csv(brazil_path)
usa_df = pd.read_csv(usa_path)
```

```python
brazil_df.info()
```

```python
usa_df.info()
```

```python
# make 6 subplots  2 rows 3 cols 
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
for i, column in enumerate(brazil_df.columns[1:]):
	sns.lineplot(x="DIM_TIME", y=column, data=brazil_df, ax=axs[i//3, i%3])
	axs[i//3, i%3].set_title(column)
```

Life Expectancy

```python

plt.figure(figsize=(9, 6))
sns.lineplot(x="DIM_TIME", y="Life Expectancy", data=brazil_df, 
             marker='8',linestyle='-.', color='orange')
plt.title("Life Expectancy in Brazil")
```

```python

plt.figure(figsize=(9, 6))
sns.lineplot(x="DIM_TIME", y="UHC_INDEX_N", data=brazil_df, linestyle='--',
             marker='D')
plt.title("UHC Index in Brazil")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a better style and palette
sns.set_theme(style="whitegrid")
palette = sns.color_palette("husl", 2)  

plt.figure(figsize=(9, 6))  
sns.lineplot(x="DIM_TIME", y="DTP3_RATE_PER_100_N", data=brazil_df, 
             marker='o', label="DTP3", color=palette[0])
sns.lineplot(x="DIM_TIME", y="MVC2_RATE_PER_100_N", data=brazil_df, 
             marker='s', label="MVC2", color=palette[1])

# Add labels and title
plt.xlabel("Time", fontsize=12)
plt.ylabel("Rate per 100", fontsize=12)
plt.title("Immunization rates over time ", fontsize=14)

# Add legend
plt.legend(title="Rate", fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(brazil_df.corr(), annot=True, cmap="coolwarm")
plt.xticks(rotation=60)
```

### Brazil vs USA

```python
# Brazil vs USA 
# Summary statistics for the USA data
usa_summary_stats = usa_df.describe()

# Correlation matrix for the USA data
usa_correlation_matrix = usa_df.corr()

# Plot comparative trends
plt.figure(figsize=(16, 10))

# Subplot 1: Life Expectancy
plt.subplot(2, 2, 1)
sns.lineplot(x='DIM_TIME', y='Life Expectancy', data=brazil_df, label='Brazil', marker="o")
sns.lineplot(x='DIM_TIME', y='Life Expectancy', data=usa_df, label='USA', marker="o")
plt.title('Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()

# Subplot 2: UHC Index
plt.subplot(2, 2, 2)
sns.lineplot(x='DIM_TIME', y='UHC_INDEX_N', data=brazil_df, label='Brazil', marker="o", color='orange')
sns.lineplot(x='DIM_TIME', y='UHC_INDEX_N', data=usa_df, label='USA', marker="o", color='purple')
plt.title('Universal Health Coverage Index Over Time')
plt.xlabel('Year')
plt.ylabel('UHC Index')
plt.legend()

# Subplot 3: Immunization Rates
plt.subplot(2, 2, 3)
sns.lineplot(x='DIM_TIME', y='DTP3_RATE_PER_100_N', data=brazil_df, label='Brazil DTP3', color='blue')
sns.lineplot(x='DIM_TIME', y='DTP3_RATE_PER_100_N', data=usa_df, label='USA DTP3', color='green')
sns.lineplot(x='DIM_TIME', y='MVC2_RATE_PER_100_N', data=brazil_df, label='Brazil MVC2', linestyle="--", color='red')
sns.lineplot(x='DIM_TIME', y='MVC2_RATE_PER_100_N', data=usa_df, label='USA MVC2', linestyle="--", color='brown')
plt.title('Immunization Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Rate per 100')
plt.legend()

```
